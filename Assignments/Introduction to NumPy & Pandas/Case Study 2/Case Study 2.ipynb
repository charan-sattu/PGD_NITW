```python
import pandas as pd
```

```python
# reading the files

ds_df = pd.read_csv('DSScoreTerm1.csv')
math_df = pd.read_csv('MathScoreTerm1.csv')
physics_df = pd.read_csv('PhysicsScoreTerm1.csv')
final_df = pd.read_csv('ScoreFinal.csv')  # for reference
```

```python
# removing 'Name' & 'Ethinicity' columns

def remove(df):
    
    return df.drop(columns = ['Name', 'Ethinicity'], axis=1)
```

```python
ds_df_1,math_df_1,physics_df_1 = remove(ds_df),remove(math_df),remove(physics_df)
```

```python
ds_df_1
```

```python
math_df_1
```

```python
physics_df_1
```

```python
# Merge all the df's - (Concat - as per the 'ScoreFinal.csv' given)

final_df = pd.concat([ds_df_1,math_df_1,physics_df_1])
```

```python
final_df
```

```python
# Checking for NA values in Score column

final_df['Score'].isna().value_counts()
```

```python
#Fill missing score data with zero or Mean of class

#final_df['Score'] = final_df['Score'].fillna(0)
final_df['Score'] = final_df['Score'].fillna(final_df['Score'].mean)
final_df['Score'].isna().value_counts()
```

```python
final_df['Sex'] = [1 if x=='M' else 2 for x in final_df['Sex']]
```

```python
final_df['Sex'].value_counts()
```

```python
final_df
```

```python
final_df.to_csv('ScoreFinal.csv')
```
